```python
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, MinMaxScaler

data = pd.read_excel('side_effect_data 1.xlsx')









```

```python
# 1. Eksik Verilerin İşlenmesi

# Kategorik veriler için SimpleImputer (mod ile doldurma)
cat_imputer = SimpleImputer(strategy='most_frequent')

# Kategorik sütunlarda eksik değerleri doldurma (örneğin 'Cinsiyet', 'Kan Grubu')
data['Cinsiyet'] = cat_imputer.fit_transform(data[['Cinsiyet']])
data['Kan Grubu'] = cat_imputer.fit_transform(data[['Kan Grubu']])

# Sayısal veriler için KNNImputer (komşu değerlerle doldurma)
num_imputer = KNNImputer(n_neighbors=5)

# Sayısal sütunlarda eksik değerleri doldurma (örneğin 'Kilo', 'Boy')
data[['Kilo', 'Boy']] = num_imputer.fit_transform(data[['Kilo', 'Boy']])
```

```python
# 2. Kategorik Verilerin Kodlanması

# 'Cinsiyet' ve 'Kan Grubu' için LabelEncoder kullanımı
le = LabelEncoder()
data['Cinsiyet'] = le.fit_transform(data['Cinsiyet'])
data['Kan Grubu'] = le.fit_transform(data['Kan Grubu'])

# Alternatif olarak OneHotEncoder kullanımı
# onehot_encoder = OneHotEncoder(sparse=False, drop='first')
# cinsiyet_encoded = onehot_encoder.fit_transform(data[['Cinsiyet']])
```

```python

# 3. Sayısal Özelliklerin Standardizasyonu veya Normalizasyonu

# StandardScaler kullanarak sayısal verilerin standardizasyonu
scaler = StandardScaler()

# 'Kilo', 'Boy' ve 'Ilac_Kullanim_Suresi' sütunlarını standardize etme
data[['Kilo', 'Boy', 'Ilac_Kullanim_Suresi']] = scaler.fit_transform(data[['Kilo', 'Boy', 'Ilac_Kullanim_Suresi']])

# Alternatif olarak MinMaxScaler ile normalizasyon
# minmax_scaler = MinMaxScaler()
# data[['Kilo', 'Boy', 'Ilac_Kullanim_Suresi']] = minmax_scaler.fit_transform(data[['Kilo', 'Boy', 'Ilac_Kullanim_Suresi']])

# Temizlenmiş ve işlenmiş veri setini görüntüleme
data.head()
```
